## Running a Histogram Using MLab in 7 bins

```python
num_bins = 7
n, bins, patches = plt.hist(Boats["Power Boats (in 1000's)"], num_bins, facecolor = 'blue', alpha = 0.9)
plt.xlabel ('Power Boats in 1000s')
plt.ylabel ('Frequency')
plt.title ('Histogram of the Number of Power Boats Listed in Florida')
plt.show()
```
